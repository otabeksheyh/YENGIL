```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv",index_col=0)
df.head()
```

```python
df.fillna(0) # nan qiymatlarga standart 0 qiymat bera olamiz
```

```python
df.fillna({'horsepower':100,'wheel-base':90}) # aynan ustunlarning nan qiymatlarini toldirmoqchi bolsak {lugat } korinishida bersak boladi
```

```python
df.fillna({'horsepower':np.mean(df['horsepower']), 'wheel-base':np.mean(df['wheel-base'])}) # ortacha qiymatni oladi nan qiymat orniga
```

```python
df
```

```python
df.fillna(method='ffill') # nan qiymatdan oldingi  qiymatni nan orniga qaytaradi
```

```python
df.fillna(method='bfill') # nan qiymatdan kngi kelgan qiymatni nan qiymat orniga qaytaradi
```

```python
df.fillna(df.mean()) # barcha nan qiymatlarni ortacha qiymatga aylanitira olamiz
```

```python
dfh = df[df['body-style']=='hatchback'] # avval hatchback rusumli mashinani chiqazib dfh ga = lab oldik
```

```python
dfh.fillna(dfh.mean) # dfh jadvalidagi ajratgan malumotlarimizni nan qiymatini ortachaga = lab olamiz
```

```python
df['num-of-cylinders'].unique()
```

```python
dict = {'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8,'twelve':12}
df['cylinders'] = df['num-of-cylinders'].map(dict)
df
```
