```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv', index_col=0)
df.head()
```

```python
df.price.describe() # min max qiymatlarni qaytaradi
```

```python
bins = [0, 10000, 20000, 50000]
```

```python
range = pd.cut(df.price, bins)
range
```

```python
range.value_counts()
```

```python
names = ["arzon", "orta", "qimmat"]
range = pd.cut(df.price, bins, labels=names)
range.value_counts()
```

```python
range = pd.cut(df.price, 3) #binsdan foydalanmasdan shunchaki sonni kiritish kifoya
range.value_counts()
```

```python
range = pd.qcut(df.price, 3) # qcut() yordamida teng taqsimlab muozanatni buzmay guruhga bola olamiz
range.value_counts()
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv', index_col=0)
df.head()
```
